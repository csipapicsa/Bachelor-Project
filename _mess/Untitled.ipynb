```python
import pandas as pd
import numpy as np
import folium  #<--- This library helps us plot data on maps
from folium.plugins import MarkerCluster
 
#Convert your data into necessary Data Types
covid = pd.read_csv("COVID19-Charges-Clean.csv")
 
covid.LAT=covid.LAT.astype("float64")
covid.LON=covid.LON.astype("float64")
 
covid.head()
```

```python
# Create a map centered on LOCATION, with a ZOOM_START level of 10
GTA = folium.Map(location=[covid.LAT.mean(), covid.LON.mean()], zoom_start=10)
```
